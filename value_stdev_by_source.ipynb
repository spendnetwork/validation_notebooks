```python
import pandas as pd
import altair as alt
```

Using data set of uk sources, excluding the main publishers, TED and Contracts Finder

```python
path = 'https://raw.githubusercontent.com/spendnetwork/validation_notebooks/master/csvs/td_uk_recent_excl_ted_cf.csv?token=ADX2I7ZGCPSAB7AAUCSCBU26YG7DC'
df = pd.read_csv(path)
```

Excluding missing and unrealistic values

```python
df1 = df[df['value'].notna()]
df1 = df1[df1['value']< 100000000]
df1 = df1[df1['value'] != 0]
```

Displaying Chart of the average value by source, with standard deviation

```python
chart = alt.Chart(df1, width=500, height=300, title="Average Value")

points = chart.mark_point(color='#91D739', filled=True, size=50).encode(
         alt.X('source'),
         alt.Y('mean(value):Q'),
        )

bars = chart.mark_errorbar(extent='stdev', color='#91D739', clip=True).encode(
          x=alt.X('source'),
          y=alt.Y('mean(value):Q')
        )

full_chart = alt.layer(
            points, bars
            ).configure_title(fontSize=18, anchor='start',
                    color= "#91D739", font='Helvetica', fontWeight=400)


full_chart
```
