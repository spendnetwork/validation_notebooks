```python
import pandas as pd
import altair as alt
```

```python
url = 'https://raw.githubusercontent.com/spendnetwork/validation_notebooks/master/csvs/ocds_bravosolution_2020.csv?token=ADX2I73QAN3UW3BENNBKWF26YG3MO'
df = pd.read_csv(url)
```

Output chart of counts by day

```python
df['pubdate'] = pd.to_datetime(df['pubdate']).dt.strftime('%Y-%m-%d')
chart = alt.Chart(df, width=900, height=300, title="Volume of Rows per Day").mark_line(color='#91D739').encode(
            alt.X('pubdate',
            axis=alt.Axis(tickCount=12, title="https://openopps.com",
            grid=False, 
            labelFont='Helvetica', 
            labelFontSize=12, 
            labelAngle=0, 
            titleAlign='left',
            titleAnchor='start',
            titleFont='Helvetica',
            titleFontSize= 11,
            titleFontWeight= 300,
            titleX= -0,
            titleY= 30)
                   ),
         alt.Y('count(*):Q',
               axis=alt.Axis(title="",
               format=".0s",)
                       ),
                )

nearest = alt.selection(type='single', nearest=True, on='mouseover',fields=['pubdate'], empty='none')
selectors = alt.Chart(df).mark_point(filled=True).encode(
            x='yearmonthdate(pubdate):O',
            opacity=alt.value(0),
            ).add_selection(
                nearest
            )
points = chart.mark_point().encode(
         opacity=alt.condition(nearest, alt.value(1), alt.value(0))
        )


text = chart.mark_text(align='left', dx=10, dy=-10).encode(
       text=alt.condition(nearest, 'count(*):Q', alt.value(' '))
        )

rules = alt.Chart(df).mark_rule(color='gray').encode(
            x='yearmonthdate(pubdate):O',
        ).transform_filter(
            nearest
        )

full_chart = alt.layer(
            chart, selectors, points, rules, text
            ).configure_view(
              strokeWidth=0
            ).configure_title(fontSize=18, anchor='start',
                    color= "#91D739", font='Helvetica', fontWeight=400)

full_chart
```
